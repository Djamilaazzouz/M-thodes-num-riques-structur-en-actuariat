### Application du modèle Lee Carter  pour les Femmes 

```python
Femme.head()
```

```python
import statsmodels.api as sm
# Transformation des données en logarithmes
femme_log = np.log(Femme.iloc[:, 1:])

# Calcul de l'âge moyen pour chaque groupe d'âge
age = np.array([0.5, 2.5, 7, 12, 17, 22, 27, 32, 37, 42, 47, 52, 57, 62, 67, 74.5, 84.5, 100])

# Définition du modèle de Lee-Carter
model = sm.OLS(femme_log.mean(axis=0), sm.add_constant(age))

# Estimation des paramètres du modèle
results = model.fit()

# Affichage des résultats de l'estimation
print(results.summary())
print("Parameters: ", results.params)
print("Standard errors: ", results.bse)
print("Predicted values: ", results.predict())
```

```python
age=pd.DataFrame(age)
age=age.rename(columns={0: 'age'})
```

```python
predict=results.predict()
predict=pd.DataFrame(predict)
prediction=pd.concat([age,predict],axis=1)
prediction = prediction.rename(columns={0: 'log_Taux_mort_predit'})
prediction.head()
```

```python
# Créer une figure de taille personnalisée
fig = plt.figure(figsize=(10, 6))

# Tracer le graphe en ligne
plt.bar(prediction['age'], prediction['log_Taux_mort_predit'],color='blue')

## Ajouter un titre et des étiquettes d'axe
plt.title('Logarithme du taux de mortalité prédit en fonction de l\'âge')
plt.xlabel('Âge')
plt.ylabel('Logarithme du taux de mortalité prédit: $\log\mu_x$')

# Afficher le graphe
plt.show()



```

```python
mortality=np.exp(predict)
mortality=mortality
mortality=pd.concat([age,mortality],axis=1)
mortality= mortality.rename(columns={0: 'Taux_mort_estimée'})
mortality.head()
```

```python
# Créer une figure de taille personnalisée
fig = plt.figure(figsize=(10, 6))

# Tracer le graphe en ligne
plt.plot(mortality['age'], mortality['Taux_mort_estimée'],color='green')

# Ajout des labels d'axe et du titre
plt.xlabel('Age')
plt.xticks(rotation=90)
plt.ylabel('Taux de mortalité : $\mu_x$')
plt.title('Taux de mortalité par age pour les Femmes')

# Affichage du graphique
plt.show()


```

```python
# Prédiction de la mortalité pour les 5 prochaines années
t = np.arange(Femme.shape[0], Femme.shape[0] + 5)
a_x = np.exp(results.params[0] + np.outer(np.ones(5), results.params[1] * age))
b_x = np.exp(np.outer(t, results.resid))

# Calcul de la mortalité prévue
femme_pred = pd.DataFrame(data=a_x * b_x, columns=Femme.columns[1:], index=t)

# Affichage des résultats de la prédiction
femme_pred
```
